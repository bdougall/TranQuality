```python
# Import the model from the cloned repository
from indicTrans.inference.engine import Model
```

```python
# Load the model
indic2en_model = Model(expdir='/content/drive/My Drive/W266_Project_Data/testing/en-indic')
```

```python
# The path to the data on my drive
D = '/content/drive/My Drive/W266_Project_Data/pmi_data'
```

```python
import csv
import numpy as np
import pandas as pd
```

```python
# Load the English dev dataset from the dev data for translation
english_eval_df = pd.read_csv(D+"/dev/dev.en", sep = "\t", header = None)
```

```python
# Give the column of text a meaningful name
english_eval_df.rename(columns={0:'input_text'}, inplace=True)
```

```python
# Translate the dev data from English to Hindi
dev_preds = pd.DataFrame(indic2en_model.batch_translate(english_eval_df.input_text.tolist(), 'en', 'hi'))

# Save the Hindi translations to a csv file
dev_preds.to_csv(D+'/predicted_text/indic_trans_hi_dev.csv')
```

```python
# Translate the dev data from English to Tamil and save the translations to a csv file
ta_dev_preds = pd.DataFrame(indic2en_model.batch_translate(english_eval_df.input_text.tolist(), 'en', 'ta'))
ta_dev_preds.to_csv(D+'/predicted_text/indic_trans_ta_dev.csv')
```

```python
# Translate the dev set from English to Malayalm and save to a csv file
ml_dev_preds = pd.DataFrame(indic2en_model.batch_translate(english_eval_df.input_text.tolist(), 'en', 'ml'))
ml_dev_preds.to_csv(D+'/predicted_text/indic_trans_ml_dev.csv')
```

```python
# Get the source and target texts for the translated Hindi dev texts
hi_actual = pd.read_csv(D+"/dev/dev.hi", sep = "\t", header = None, quoting=csv.QUOTE_NONE)
hi_pred = pd.read_csv(D+"/predicted_text/IndicTrans/dev/indic_trans_hi_dev.csv")
hi_pred.drop(columns={"Unnamed: 0"}, inplace=True)
# Save the predicted values as a text file
np.savetxt("indictrans_hi.txt", hi_pred.values, fmt = "%s")
# Save the actual values as a text file
np.savetxt("dev_hi.txt", hi_actual.values, fmt = "%s")
```

```python
ml_actual = pd.read_csv(D+"/dev/dev.ml", sep = "\t", header = None, quoting=csv.QUOTE_NONE)
ml_pred = pd.read_csv(D+"/predicted_text/IndicTrans/dev/indic_trans_ml_dev.csv")
ml_pred.drop(columns={"Unnamed: 0"}, inplace=True)
# Save the predicted values as a text file
np.savetxt("indictrans_ml.txt", ml_pred.values, fmt = "%s")
# Save the actual values as a text file
np.savetxt("dev_ml.txt", ml_actual.values, fmt = "%s")
```

```python
ta_actual = pd.read_csv(D+"/dev/dev.ta", sep = "\t", header = None, quoting=csv.QUOTE_NONE)
ta_pred = pd.read_csv(D+"/predicted_text/IndicTrans/dev/indic_trans_ta_dev.csv")
ta_pred.drop(columns={"Unnamed: 0"}, inplace=True)
# Save the predicted values as a text file
np.savetxt("indictrans_ta.txt", ta_pred.values, fmt = "%s")
# Save the actual values as a text file
np.savetxt("dev_ta.txt", ta_actual.values, fmt = "%s")
```
